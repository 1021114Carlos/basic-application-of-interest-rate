```python
# When interest rates compounds in your favor.

monthly_amount = float(input("amount to save every month $: " ))
print("\n")
apr = float(input("What is the annualized interest rate (e.g. 5%): "))/100
print("\n")
time_horizon = float(input("How many years are you going to be saving (e.g. 1, 2): "))
print("\n")
fv = 0

for month in range(1, 12 + 1):
  monthly_savings = monthly_amount + fv
  n = 365
  t = time_horizon
  updated_monthly_savings = monthly_savings*(1 + (apr/n))**(n*t)
  fv = updated_monthly_savings
print(f"Future value of your savings after {time_horizon} year(s) is ${fv:.6f}")
```
